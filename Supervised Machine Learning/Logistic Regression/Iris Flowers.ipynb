```python
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
x = iris.data[:,:2]
y = (iris.target !=0 ) * 1
```

```python
plt.figure(figsize=(6,6))
plt.scatter(x[y==0][:,0], x[y==0][: , 1], color='g',label ='0')
plt.scatter(x[y==1][:,0], x[y==1][: , 1], color='y',label ='1')
plt.legend()
```
