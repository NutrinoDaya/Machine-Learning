```python
from statistics import mean 
import numpy as np 
import matplotlib.pyplot as plt 
from matplotlib import style 
style.use('fivethirtyeight')
```

```python
xs = np.array([1,2,3,4,5,6], dtype=np.float64)
ys = np.array([5,4,6,5,6,7], dtype=np.float64)
plt.scatter(xs,ys)
plt.show()
```

```python
def best_fit_slope(xs,ys) : 
    m =  ( (mean(xs)*mean(ys)) - mean(xs*ys) ) / ((mean(xs)**2) - mean(xs**2))
    b = mean(ys) - m*mean(xs)

    return m ,b

m ,b = best_fit_slope(xs,ys)
print(m,b)
```

```python
regression_line = [(m*x)+b for x in xs ]
predict_x = 8 
predict_y = (m*predict_x) + b

plt.scatter(xs,ys)
plt.scatter(predict_x,predict_y,color='g',s=40)
plt.plot(regression_line)
plt.show()
```

```python
# How good of a fit is the best fit line ? 
# we can use the sqaured error 
#r^2 = 1 - SE y hat / SE mean(y)
def squared_error(ys_origi,ys_line):

    return sum((ys_line - ys_origi)**2)
def coefficient_of_determination(ys_orig,ys_line):
    y_mean_line = [mean(ys_orig) for y in ys_orig]
    squared_err_reg = squared_error(ys_orig,ys_line)
    squared_err_reg_y_mean = squared_error(ys_orig,y_mean_line)
    return 1 -(squared_err_reg/squared_err_reg_y_mean)
r_squared = coefficient_of_determination(ys,regression_line)
print(r_squared)
```
