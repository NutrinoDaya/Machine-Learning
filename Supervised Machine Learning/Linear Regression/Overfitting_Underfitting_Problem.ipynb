Underfitting

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
```

```python
#Generate non-linear data 
np.random.seed(0)
X = np.sort(5*np.random.rand(80,1),axis =0)
y = np.sin(X).ravel() + np.random.normal(0,0.1,X.shape[0])
```

```python
x_train , x_test = X[:40] , X[40:]
y_train,y_test = y[:40] , y[40:]
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_train_pred = model.predict(x_train)
y_test_pred = model.predict(x_test)
```

```python
mse_train = mean_squared_error(y_train,y_train_pred)
mse_test = mean_squared_error(y_test,y_test_pred)
```

```python

# Visualize the results
plt.scatter(X, y, s=20, edgecolor="black", c="darkorange", label="data")
plt.plot(X, model.predict(X), color="cornflowerblue", linewidth=2, label="linear fit")
plt.xlabel("Feature (X)")
plt.ylabel("Target (y)")
plt.title("Linear Regression (Underfitting)")
plt.legend()
plt.show()
```

Overfitting 

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
```

```python

# Generate simple data
np.random.seed(0)
X = np.random.rand(80, 1)
y = 1 + 2 * X.ravel() + np.random.normal(0, 0.1, X.shape[0])
```

```python

# Split the data into training and testing sets
X_train, X_test = X[:40], X[40:]
y_train, y_test = y[:40], y[40:]
```

```python

# Create a high-degree polynomial model
poly = PolynomialFeatures(degree=15)
X_train_poly = poly.fit_transform(X_train)
```

```python
model = LinearRegression()
# Fit the polynomial model
model.fit(X_train_poly, y_train)
```

```python

# Transform test data and make predictions
X_test_poly = poly.transform(X_test)
y_test_pred = model.predict(X_test_poly)
```

```python

# Calculate MSE for testing data
mse_test = mean_squared_error(y_test, y_test_pred)
temp = np.linspace(0, 1, 100)
print('temp')
print(temp)
temp2 = np.linspace(0, 1, 100)[:, np.newaxis]
print('temp2')
print(temp2)
# Visualize the results
X_range = np.linspace(0, 1, 100)[:, np.newaxis]
X_range_poly = poly.transform(X_range)
y_range_pred = model.predict(X_range_poly)
```

```python

plt.scatter(X, y, s=20, edgecolor="black", c="darkorange", label="data")
plt.plot(X_range, y_range_pred, color="cornflowerblue", linewidth=2, label="polynomial fit (degree=15)")
plt.xlabel("Feature (X)")
plt.ylabel("Target (y)")
plt.title("Polynomial Regression (Overfitting)")
plt.legend()
plt.show()
```
