```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python

# Sample data
temperature = np.array([15, 20, 25, 30, 35, 40])
ice_cream_sales = np.array([50, 75, 120, 160, 200, 220])
```

```python

# Reshape the data to a 2D array
X = temperature.reshape(-1, 1)
```

```python

# Fit a linear regression model to visualize the linear fit
lin_reg = LinearRegression()
lin_reg.fit(X, ice_cream_sales)
```

```python

# Predictions from the linear model
y_pred_lin = lin_reg.predict(X)
```

```python

# Fit a polynomial regression model
poly_features = PolynomialFeatures(degree=4)
X_poly = poly_features.fit_transform(X)
poly_reg = LinearRegression()
poly_reg.fit(X_poly, ice_cream_sales)
```

```python

# Predictions from the polynomial model
y_pred_poly = poly_reg.predict(X_poly)
```

```python

# Visualize the data and regression curves
plt.scatter(temperature, ice_cream_sales, label='Data')
plt.plot(temperature, y_pred_lin, color='r', label='Linear Regression')
plt.plot(temperature, y_pred_poly, color='g', label='Polynomial Regression (degree=2)')
plt.xlabel('Temperature (°C)')
plt.ylabel('Ice Cream Sales (units)')
plt.title('Ice Cream Sales Prediction with Polynomial Regression')
plt.legend()
plt.grid(True)
plt.show()
```
